```python
import numpy as np
from sklearn.linear_model import LinearRegression

#Data

x = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([35,45,50,60,70])

model = LinearRegression()
model.fit(x,y)
print(f"Predict salary for 6 years {model.predict([[6]])[0]}")
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
x = np.array([
    [1000,2],
    [1500 ,3],
    [2000 ,4],
    [2500 ,4],
])

y = np.array([50,70,90,120])

model = LinearRegression()
model.fit(x,y)
print(model.predict([[1000,3]])[0])
```
