```python
import pandas as pd

data = {
    "hours_studied": [2, 4, 6, 8, 10, 12, 14, 16],
    "marks": [35, 45, 55, 65, 75, 85, 90, 95],
    "attendance": [60, 65, 70, 75, 80, 85, 92, 96],
    "sleep_hours": [5, 6, 6.5, 7, 7.5, 8, 8, 8.5]
}

df = pd.DataFrame(data)
df.head()
```

```python
from matplotlib import pyplot as plt

plt.figure()
plt.plot(df['hours_studied'], df['marks'], marker='o')
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.figure()
plt.plot(df.index, df['attendance'])
plt.xlabel("Student Index")
plt.ylabel("Attendance")
plt.title("Attendance of Students")
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.bar(df.index, df['attendance'])
plt.xlabel("Student Index")
plt.ylabel("Attendance")
plt.title("Attendance of Students")
# plt.grid(True)
plt.show()
```

```python
corr_mtrx = df.corr()
print(corr_mtrx)
```

```python
import seaborn as sns

plt.figure()
sns.heatmap(corr_mtrx, annot=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
